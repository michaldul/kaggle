```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
%matplotlib inline
```

```python
events = pd.read_csv('e:/data/events.csv')
```

```python
events['datetime'] = events['timestamp'].apply(
    lambda ts: datetime.datetime.fromtimestamp((ts + 1465876799998)/1000))
events['date'] = events['datetime'].apply(lambda dt: dt.date())
```

```python
plt.plot(events.groupby('date').count()['display_id'])
plt.xticks(rotation='vertical')
plt.show()
```
