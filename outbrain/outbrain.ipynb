```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
%matplotlib inline
```

```python
events = pd.read_csv('./data/events.csv')
```

```python
events['datetime'] = events['timestamp'].apply(
    lambda ts: datetime.datetime.fromtimestamp((ts + 1465876799998)/1000))
events['date'] = events['datetime'].apply(lambda dt: dt.date())
```

```python
events['date_hour'] = events['timestamp'].apply(
    lambda ts: datetime.datetime.fromtimestamp((ts + 1465876799998)/1000).replace(minute=0, second=0, microsecond=0))
```

```python
plt.figure(figsize=(18, 6))
plt.plot(events.groupby('date_hour').count()['display_id'])
plt.xticks(rotation='vertical')
plt.show()
```
